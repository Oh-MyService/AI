```python
# 필요한 라이브러리 임포트
import torch
from diffusers import StableDiffusionPipeline
import os
from PIL import Image

# 모델 파이프라인 설정 함수
def prepare_pipeline(model_name):
    # Stable Diffusion 모델을 불러옵니다
    pipeline = StableDiffusionPipeline.from_pretrained(
        model_name, 
        torch_dtype=torch.float16  # float16 사용으로 GPU 메모리 효율화
    ).to('cuda')
    return pipeline

# 이미지 생성 함수 정의
def generate_images(prompt, neg_prompt, width, height, num_inference_steps, guidance_scale, seed, num_images):
    # 시드 설정으로 결과 재현 가능하게 설정
    generator = torch.Generator(device='cuda').manual_seed(seed)

    # 이미지 생성
    images = pipeline(
        prompt=prompt,
        negative_prompt=neg_prompt,
        width=width,
        height=height,
        num_inference_steps=num_inference_steps,
        guidance_scale=guidance_scale,
        num_images_per_prompt=num_images,
        generator=generator
    )
    
    return images

# 모델 초기화 (모델 경로를 SD 모델의 경로로 설정)
model_path = "C:\\Users\\ParkChunSoo\\Downloads\\floral_lora.safetensors"  # SD 모델 이름 또는 경로
pipeline = prepare_pipeline(model_path)

# 테스트할 파라미터 설정
test_prompt = "A repeating pattern of cat, seamless fabric textile design, masterpiece."
neg_prompt = "aaaaaaaaa"
options = {
    "width": 1024,
    "height": 1024,
    "sampling_steps": 20,
    "cfg_scale": 7,
    "seed": 42
}

# 이미지 생성 및 표시
try:
    images = generate_images(
        prompt=test_prompt,
        neg_prompt=neg_prompt,
        width=options["width"],
        height=options["height"],
        num_inference_steps=options["sampling_steps"],
        guidance_scale=options["cfg_scale"],
        seed=options["seed"],
        num_images=4  # 4개의 이미지를 생성
    )
    
    # 생성된 이미지 표시
    for idx, image in enumerate(images):
        display(image)  # Jupyter Notebook에 이미지 표시

except Exception as e:
    print(f"An error occurred: {e}")
```
